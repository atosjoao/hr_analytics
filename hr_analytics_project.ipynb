### **3.2 Por Setor**
1. Qual setor possui mais demiss√µes?
2. Qual setor com mais pessoas?
3. Em qual setor est√° registrada a melhor m√©dia de satisfa√ß√£o? E a pior?
4. Em qual setor h√° mais pessoas com 100% de satisfa√ß√£o?
5. Em qual setor as pessoas mais melhoraram suas avalia√ß√µes em rela√ß√£o as anteriores? E o pior?
6. Em m√©dia, qual setor acumula mais horas trabalhadas mensalmente?
7. Em m√©dia, qual setor possui os trabalhadores mais longevos? E qual o setor mais rotativo?
8. Qual o setor mais perigoso da empresa?
9. Qual setor possui a maior taxa de promo√ß√µes?

```python
# 1. Qual setor possui mais demiss√µes?
df1 = (df.loc[df['left']==1, ['Department','left']].groupby('Department')
                                                   .count().sort_values(by='left',ascending=False)
                                                   .reset_index())
print(f'O setor que mais houve demiss√µes foi o setor de {df1["Department"][0]}, com {df1["left"][0]} demiss√µes.\n'
      f'Isso representa {(df1["left"][0]/df1["left"].sum())*100:,.2f}% do total de demiss√µes.')
```

```python
# 2. Qual setor com mais pessoas?
df1 = df.loc[:, ['Department','Emp_Id']].groupby('Department').count().sort_values(by='Emp_Id',ascending=False).reset_index()
print(f'O setor com mais pessoas √© o de {df1["Department"][0]}, com {df1["Emp_Id"][0]} colaboradores. \n'
      f'Isso corresponde a {(df1["Emp_Id"][0]/df["Emp_Id"].nunique()):,.2f}% do total de colaboradores.')
```

```python
# 3. Em qual setor est√° registrada a melhor m√©dia de satisfa√ß√£o? E a pior?
df1 = (df.loc[:, ['Department','satisfaction_level']].groupby('Department').mean().sort_values(by='satisfaction_level',ascending=False).reset_index())
print(f'O setor com melhor m√©dia de satisfa√ß√£o √© o de {df1["Department"][0]}, com {(df1["satisfaction_level"][0])*100:,.2f}% de satisfa√ß√£o dos colaboradores. \n'
      f'J√° a pior m√©dia, fica com {df1["Department"].iloc[-1]}, com {(df1["satisfaction_level"].min())*100:,.2f}% de satisfa√ß√£o.')
```

```python
# 4. Em qual setor h√° mais pessoas com 100% de satisfa√ß√£o?
df1 = (df.loc[df['satisfaction_level']==1, ['Department','satisfaction_level']].groupby('Department')
                                                                               .count()
                                                                               .sort_values(by='satisfaction_level',ascending=False)
                                                                               .reset_index())
print(f'O setor com mais colaboradoressatisfeitos √© o de {df1["Department"][0]},' 
      f' com {(df1["satisfaction_level"][0])} notas m√°ximas de satisfa√ß√£o dos colaboradores. \n'
      f'J√° o pior, fica com {df1["Department"].iloc[-1]}, com {(df1["satisfaction_level"].iloc[-1])} notas m√°ximas.')
```

```python
# 4. Em qual setor as pessoas mais melhoraram suas avalia√ß√µes em rela√ß√£o as anteriores? E o pior?
df1 = (df.loc[:, ['Department','diff_evaluation']].groupby('Department').mean()
                                                                        .sort_values(by='diff_evaluation',ascending=False)
                                                                        .reset_index())
print(f'O setor em que as avalia√ß√µes de satisfa√ß√£o tiveram uma melhora em rela√ß√£o a anterior √© o de {df1["Department"][0]},'
      f' com {(df1["diff_evaluation"][0])*100:,.2f}% de satisfa√ß√£o dos colaboradores. \n'
      f'J√° o que mais diminuiu, foi o de {df1["Department"].iloc[-1]}, com {(df1["diff_evaluation"].min())*100:,.2f}% de satisfa√ß√£o.')
```

```python
# 5. Em m√©dia, qual setor acumula mais horas trabalhadas mensalmente?
df1 = df.loc[:, ['Department','average_montly_hours']].groupby('Department').mean().sort_values(by='average_montly_hours',ascending=False).reset_index()
print(f'Na m√©dia, o setor em que acumulou mais horas trabalhadas mensalmente √© o de {df1["Department"][0]}, com {(df1["average_montly_hours"][0]):,.2f} horas. \n'
      f'J√° o que mais diminuiu, foi o de {df1["Department"].iloc[-1]}, com {(df1["average_montly_hours"].iloc[-1]):,.2f} de horas trabalhadas.')
```

```python
# 6. Em m√©dia, qual setor possui os trabalhadores mais longevos? E qual o setor mais rotativo?
df1 = df.loc[df['left']==1, ['Department','time_spend_company']].groupby('Department').mean().sort_values(by='time_spend_company',ascending=False).reset_index()
print(f'Na m√©dia, o setor em que se tem as pessoas mais longevas √© o de {df1["Department"][0]}, com {(df1["time_spend_company"][0]):,.1f} anos. \n'
      f'J√° o que mais rotativo, √© o de {df1["Department"].iloc[-1]}, com {(df1["time_spend_company"].iloc[-1]):,.1f} de anos de empresa.')
```

```python
# 7. Qual o setor mais perigoso da empresa?
df1 = df.loc[:, ['Department','Work_accident']].groupby('Department').count().sort_values(by='Work_accident',ascending=False).reset_index()
print(f'O setor em que se tem mais acidentes √© o de {df1["Department"][0]}, com {(df1["Work_accident"][0])} acidentes registrados. \n'
      f'J√° o que mais seguro, √© o de {df1["Department"].iloc[-1]}, com {(df1["Work_accident"].iloc[-1])} acidentes registrados.')
```

```python
# 8. Qual setor possui a maior taxa de promo√ß√µes?
df1 = df.loc[df['promotion_last_5years']==1, ['Department','promotion_last_5years']].groupby('Department').count().sort_values(by='promotion_last_5years',ascending=False).reset_index()
print(f'O setor em que h√° mais promo√ß√µes √© o de {df1["Department"][0]}, com {(df1["promotion_last_5years"][0]/df1["promotion_last_5years"].sum())*100:,.2f}% das promo√ß√µes. \n'
      f'J√° o que menos promove, √© o de {df1["Department"].iloc[-1]}, com {(df1["promotion_last_5years"].iloc[-1]/df1["promotion_last_5years"].sum())*100:,.2f}% das promo√ß√µes.')
```

## **4. Estat√≠stica Descritiva**

```python
stats = df.describe().T
stats['coef_var'] = stats['std'] / stats['mean']
stats
```

```python
# Avaliando a coluna de 'satisfaction_level'
df_satisf = (df.loc[:, ['Department','satisfaction_level']].groupby(['Department'])
                                                     .agg({'satisfaction_level':['mean','median','std','count']})
                                                     .reset_index())
df_satisf['satisfaction_level',   'coef_var'] = df_satisf['satisfaction_level',   'std'] / df_satisf['satisfaction_level',   'mean']
df_satisf
```

```python
# Avaliando a coluna de Satisfa√ß√£o
df_last = (df.loc[:, ['Department','last_evaluation']].groupby(['Department'])
                                                     .agg({'last_evaluation':['mean','median','std','count']})
                                                     .reset_index())
df_last['last_evaluation',   'coef_var'] = df_last['last_evaluation',   'std'] / df_last['last_evaluation',   'mean']
df_last
```

```python
# An√°lise da coluna 'average_montly_hours' por setor
df_month = df.loc[:, ['Department','average_montly_hours']].groupby(['Department']).describe().reset_index()
df_month['average_montly_hours',   'coef_var'] = df_month['average_montly_hours',   'std'] / df_month['average_montly_hours',   'mean']
df_month
```

```python
# An√°lise da coluna 'diff_evaluation' por setor
df_dif = df.loc[:, ['Department','diff_evaluation']].groupby(['Department']).describe().reset_index()
df_dif['diff_evaluation',   'coef_var'] = df_dif['diff_evaluation',   'std'] / df_dif['diff_evaluation',   'mean']
df_dif
```

```python
# An√°lise da coluna 'time_spend_company' por setor
df_time = df.loc[df['left']==1, ['Department','time_spend_company']].groupby(['Department']).describe().reset_index()
df_time['time_spend_company','coef_var'] = df_time['time_spend_company','std'] / df_time['time_spend_company','mean']
display(df_time)
```

```python
# An√°lise da coluna 'Work_accident' por setor
df_accident = (df.loc[df['Work_accident']==1, ['Department','Work_accident']].groupby(['Department'])
                                                                             .count()
                                                                             .sort_values(by='Work_accident', ascending=True)
                                                                             .reset_index())
df_accident['percent_colaboradores(%)'] = round(((df_accident['Work_accident'] / df['Emp_Id'].count())*100),2)
df_accident['percent_total(%)'] = round(((df_accident['Work_accident'] / df_accident['Work_accident'].sum())*100),2)
df_accident
```

```python
df_promo = (df.loc[df['promotion_last_5years']==1, ['Department','promotion_last_5years']].groupby(['Department'])
                                                                             .count()
                                                                             .sort_values(by='promotion_last_5years', ascending=True)
                                                                             .reset_index())
df_promo['percent_colaboradores(%)'] = round(((df_promo['promotion_last_5years'] / df['Emp_Id'].count())*100),2)
df_promo['percent_total(%)'] = round(((df_promo['promotion_last_5years'] / df_promo['promotion_last_5years'].sum())*100),2)
df_promo
```
