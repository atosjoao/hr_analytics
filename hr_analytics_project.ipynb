```python
# 1. Qual setor possui mais demissões?
df1 = (df.loc[df['left']==1, ['Department','left']].groupby('Department')
                                                   .count().sort_values(by='left',ascending=False)
                                                   .reset_index())
print(f'O setor que mais houve demissões foi o setor de {df1["Department"][0]}, com {df1["left"][0]} demissões.\n'
      f'Isso representa {(df1["left"][0]/df1["left"].sum())*100:,.2f}% do total de demissões.')
```

```python
# 2. Qual setor com mais pessoas?
df1 = df.loc[:, ['Department','Emp_Id']].groupby('Department').count().sort_values(by='Emp_Id',ascending=False).reset_index()
print(f'O setor com mais pessoas é o de {df1["Department"][0]}, com {df1["Emp_Id"][0]} colaboradores. \n'
      f'Isso corresponde a {(df1["Emp_Id"][0]/df["Emp_Id"].nunique()):,.2f}% do total de colaboradores.')
```

```python
# 3. Em qual setor está registrada a melhor média de satisfação? E a pior?
df1 = (df.loc[:, ['Department','satisfaction_level']].groupby('Department').mean().sort_values(by='satisfaction_level',ascending=False).reset_index())
print(f'O setor com melhor média de satisfação é o de {df1["Department"][0]}, com {(df1["satisfaction_level"][0])*100:,.2f}% de satisfação dos colaboradores. \n'
      f'Já a pior média, fica com {df1["Department"].iloc[-1]}, com {(df1["satisfaction_level"].min())*100:,.2f}% de satisfação.')
```

```python
# 4. Em qual setor há mais pessoas com 100% de satisfação?
df1 = (df.loc[df['satisfaction_level']==1, ['Department','satisfaction_level']].groupby('Department')
                                                                               .count()
                                                                               .sort_values(by='satisfaction_level',ascending=False)
                                                                               .reset_index())
print(f'O setor com mais colaboradoressatisfeitos é o de {df1["Department"][0]},' 
      f' com {(df1["satisfaction_level"][0])} notas máximas de satisfação dos colaboradores. \n'
      f'Já o pior, fica com {df1["Department"].iloc[-1]}, com {(df1["satisfaction_level"].iloc[-1])} notas máximas.')
```

```python
# 4. Em qual setor as pessoas mais melhoraram suas avaliações em relação as anteriores? E o pior?
df1 = (df.loc[:, ['Department','diff_evaluation']].groupby('Department').mean()
                                                                        .sort_values(by='diff_evaluation',ascending=False)
                                                                        .reset_index())
print(f'O setor em que as avaliações de satisfação tiveram uma melhora em relação a anterior é o de {df1["Department"][0]},'
      f' com {(df1["diff_evaluation"][0])*100:,.2f}% de satisfação dos colaboradores. \n'
      f'Já o que mais diminuiu, foi o de {df1["Department"].iloc[-1]}, com {(df1["diff_evaluation"].min())*100:,.2f}% de satisfação.')
```

```python
# 5. Em média, qual setor acumula mais horas trabalhadas mensalmente?
df1 = df.loc[:, ['Department','average_montly_hours']].groupby('Department').mean().sort_values(by='average_montly_hours',ascending=False).reset_index()
print(f'Na média, o setor em que acumulou mais horas trabalhadas mensalmente é o de {df1["Department"][0]}, com {(df1["average_montly_hours"][0]):,.2f} horas. \n'
      f'Já o que mais diminuiu, foi o de {df1["Department"].iloc[-1]}, com {(df1["average_montly_hours"].iloc[-1]):,.2f} de horas trabalhadas.')
```

```python
# 6. Em média, qual setor possui os trabalhadores mais longevos? E qual o setor mais rotativo?
df1 = df.loc[df['left']==1, ['Department','time_spend_company']].groupby('Department').mean().sort_values(by='time_spend_company',ascending=False).reset_index()
print(f'Na média, o setor em que se tem as pessoas mais longevas é o de {df1["Department"][0]}, com {(df1["time_spend_company"][0]):,.1f} anos. \n'
      f'Já o que mais rotativo, é o de {df1["Department"].iloc[-1]}, com {(df1["time_spend_company"].iloc[-1]):,.1f} de anos de empresa.')
```

```python
# 7. Qual o setor mais perigoso da empresa?
df1 = df.loc[:, ['Department','Work_accident']].groupby('Department').count().sort_values(by='Work_accident',ascending=False).reset_index()
print(f'O setor em que se tem mais acidentes é o de {df1["Department"][0]}, com {(df1["Work_accident"][0])} acidentes registrados. \n'
      f'Já o que mais seguro, é o de {df1["Department"].iloc[-1]}, com {(df1["Work_accident"].iloc[-1])} acidentes registrados.')
```

## **4. Estatística Descritiva**

```python
stats = df.describe().T
stats['coef_var'] = stats['std'] / stats['mean']
stats
```

```python
# Avaliando a coluna de 'satisfaction_level'
df_satisf = (df.loc[:, ['Department','satisfaction_level']].groupby(['Department'])
                                                     .agg({'satisfaction_level':['mean','median','std','count']})
                                                     .reset_index())
df_satisf['satisfaction_level',   'coef_var'] = df_satisf['satisfaction_level',   'std'] / df_satisf['satisfaction_level',   'mean']
df_satisf
```

```python
# Avaliando a coluna de Satisfação
df_last = (df.loc[:, ['Department','last_evaluation']].groupby(['Department'])
                                                     .agg({'last_evaluation':['mean','median','std','count']})
                                                     .reset_index())
df_last['last_evaluation',   'coef_var'] = df_last['last_evaluation',   'std'] / df_last['last_evaluation',   'mean']
df_last
```

```python
# Análise da coluna 'average_montly_hours' por setor
df_month = df.loc[:, ['Department','average_montly_hours']].groupby(['Department']).describe().reset_index()
df_month['average_montly_hours',   'coef_var'] = df_month['average_montly_hours',   'std'] / df_month['average_montly_hours',   'mean']
df_month
```

```python
# Análise da coluna 'diff_evaluation' por setor
df_dif = df.loc[:, ['Department','diff_evaluation']].groupby(['Department']).describe().reset_index()
df_dif['diff_evaluation',   'coef_var'] = df_dif['diff_evaluation',   'std'] / df_dif['diff_evaluation',   'mean']
df_dif
```

```python
# Análise da coluna 'time_spend_company' por setor
df_time = df.loc[df['left']==1, ['Department','time_spend_company']].groupby(['Department']).describe().reset_index()
df_time['time_spend_company','coef_var'] = df_time['time_spend_company','std'] / df_time['time_spend_company','mean']
display(df_time)
```

```python
# Análise da coluna 'Work_accident' por setor
df_accident = (df.loc[df['Work_accident']==1, ['Department','Work_accident']].groupby(['Department'])
                                                                             .count()
                                                                             .sort_values(by='Work_accident', ascending=True)
                                                                             .reset_index())
df_accident['percent_colaboradores(%)'] = round(((df_accident['Work_accident'] / df['Emp_Id'].count())*100),2)
df_accident['percent_total(%)'] = round(((df_accident['Work_accident'] / df_accident['Work_accident'].sum())*100),2)
df_accident
```

```python
df_promo = (df.loc[df['promotion_last_5years']==1, ['Department','promotion_last_5years']].groupby(['Department'])
                                                                             .count()
                                                                             .sort_values(by='promotion_last_5years', ascending=True)
                                                                             .reset_index())
df_promo['percent_colaboradores(%)'] = round(((df_promo['promotion_last_5years'] / df['Emp_Id'].count())*100),2)
df_promo['percent_total(%)'] = round(((df_promo['promotion_last_5years'] / df_promo['promotion_last_5years'].sum())*100),2)
df_promo
```

## **5. Conclusão**
##### **1.** O nível de satisfação como um todo caiu quase 11% desde a última avaliação e não apenas isso, as respostas estão menos homogêneas, ou seja, há menos consenso e unidade nessa categoria, com uma amplitude maior nas respostas. Isso se dá, não apenas pelo aumento no desvio padrão e o coeficiente de variação, como também se observarmos que na última avaliação, a resposta mais baixa foi 36% de satisfação, agora, a mínima é de 9%, um aumento de 4x na amplitude total das respostas.<br>
##### **2.** Todos os setores apresentaram queda na satisfação. O setor de **Accounting** foi o que mais caiu, com média de 13% de queda em relação a avaliação anterior, com respostas chegando a 87% abaixo da anterior. O que menos teve alteração foi o de **Product_mng**, com redução de apenas 9%, mas que também alcançou picos de 87% de diferença.<br>
##### **3.** Os colaboradores do setor de **Management** não apenas são os que, em média, trabalham mais horas, como seus resultados são os mais homogêneos, ou seja, a média de cada trabalhador próxima da média do setor. Além disso, é o setor com menos pessoal, o que o coloca como o setor com os dados mais consistentes nesta categoria.
##### **4.** Das pessoas que se desligaram, a média de permanência até o desligamento vai de 3.8 a 4 anos, com tempo mínimo de 2 anos e máximo de 6. Sendo o **RH** o setor mais rotativo e o **RandD** o mai longevo.
##### **5.** **Sales**, **Technical** e **Supprt** possuem as maiores taxas de acidentes entre os setores. Juntos, somam quase 60% do total de acidentes. Sales é o mais perigoso, de cada 100 colaboradores, três sofreram algum tipo de acidente de trabalho.
##### **6.** **Product_mng**, é um caso curioso. É o único setor em que não houve nenhuma promoção nos últimos 5 anos, mesmo assim, possui a menor variação de avaliação e uma das maiores médias de satisfação entre os colaboradores. O setor campeão de promoções é o de **Sales**, de cada 10 promoções, 3 são para colaboradores dessa área.
##### **7.** O setor de vendas, é o que mais demitiu dentre todos na base de dados. O perfil médio de um colaborador que se desliga, possui 43% de satisfação, 3.8 anos de empresa, um decréscimo de -27.99% em relação a última avaliação e possui uma jornada de 207 horas mensais, ou 9,85h por dia.
